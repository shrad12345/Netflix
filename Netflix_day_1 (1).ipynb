```python
data_frame_cleaned['age_certification'] = data_frame_cleaned['age_certification'].fillna('NR')
```

```python
import ast
```

```python
data_frame_cleaned['genres'] = data_frame_cleaned['genres'].apply(ast.literal_eval)
```

```python
data_frame_cleaned['production_countries'] = data_frame_cleaned['production_countries'].apply(ast.literal_eval)
```

```python
data_frame_cleaned['seasons'] = data_frame_cleaned['seasons'].fillna(0).astype(int)
```

```python
num_cols = ['imdb_score', 'imdb_votes', 'tmdb_popularity', 'tmdb_score']
for col in num_cols:
    data_frame_cleaned[col] = data_frame_cleaned[col].fillna(data_frame_cleaned[col].mean())
```
